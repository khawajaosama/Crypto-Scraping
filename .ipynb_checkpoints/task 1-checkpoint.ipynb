# Task No 3:

###  Discard the cryptocurrencies without a market capitalization

```python
dec_6.query('market_cap_usd>0',inplace=True)
dec_6
```

# Task No 4:

## Compairing Bitcoin compared with the rest of the cryptocurrencies

```python
TOP_CAP_TITLE = 'Top 10 market capitalization'
TOP_CAP_YLABEL = '% of total cap'

x_axis=np.array((dec_6['id'].head(10)).values)
capital=(dec_6['market_cap_usd'].head(10)).values
print(capital)
x_axis
```

```python
y_axis=(capital/sum(capital))*100
```

```python
y_axis
raw=np.arange(0,10)
raw
```

```python
plt.bar(raw,y_axis)#color=['orange', 'green', 'orange', 'cyan', 'cyan', 'blue', 'silver', 'orange', 'red', 'green'])
plt.xticks(fontsize=10,rotation=45)
plt.xlabel(TOP_CAP_TITLE)
plt.ylabel(TOP_CAP_YLABEL)

plt.show()
```

```python
#def top10_subplot(volatility_series, title):
fig, sets = plt.subplots(nrows=1,ncols=2,figsize(12,8))
set_1 =dec_6[['id','percent_change_24h']]
set_1[['percent_change_24h']]=set_1[['percent_change_24h']].astype(dtype='f')
set_1=set_1.sort_values('percent_change_24h',ascending=True)
set_1=set_1.head(10)


x_labels=set_1['id'].values
x_range=np.arange(len(x_labels))
plt.xticks(x_range,list(x_labels),rotation=90)

y_range=set_1['percent_change_24h'].values
sets[1].set_title("Winers", fontsize=15, color='r')
sets[1].set_ylabel("Market Capital USD", fontsize=15)
sets[1].bar(x_range,y_range)
```
