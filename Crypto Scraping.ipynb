## Compairing Bitcoin compared with the rest of the cryptocurrencies

#### Storing the first 10 ids in 'x_axis' and market capital in 'cap'

```python
TOP_CAP_TITLE = 'Top 10 market capitalization'
TOP_CAP_YLABEL = '% of total cap'

x_axis=without_cap['id'].head(10)
capital=without_cap['market_cap_usd'].head(10)
print(capital)
x_axis
```

```python
raw=np.arange(10)
plt.bar(raw,y_axis)
plt.xticks(raw,x_axis,fontsize=10,rotation=45)
plt.xlabel(TOP_CAP_TITLE)
plt.ylabel(TOP_CAP_YLABEL)

plt.show()
```

# Task No 7:

## Well, we can already see that things are a bit crazy

```python
def sub_plots(volatility,title):
    
    figure, axes =plt.subplots(nrows=1,ncols=2,figsize=(8,6))
    
    volatility=volatility[['id','percent_change_24h']].dropna().sort_values(by='percent_change_24h')
    list_of_10=np.arange(10)
    #For 1st bar chart
    
    lossers=volatility.head(10)
    
    axes[0].bar(list_of_10,lossers.percent_change_24h) #plotting bar using id and % change
    axes[0].set_xticks(list_of_10)                     #setting xticks
    axes[0].set_xticklabels(lossers.id)                #now setting labels on x ticks
    axes[0].set_title("Losers",color='red')            #setting title
    axes[0].set_ylabel('% change')                     #setting y label
    axes[0].tick_params(rotation=90)                   #rotating the x-axis labels at 90*
    
    #For 2nd bar chart
    
    winners=volatility.tail(10)
    
    axes[1].bar(list_of_10,winners.percent_change_24h) #plotting bar using id and % change
    axes[1].set_xticks(list_of_10)                     #setting xticks
    axes[1].set_xticklabels(winners.id)                #now setting labels on x ticks
    axes[1].set_title("Winners",color='red')           #setting title
    axes[1].set_ylabel('% change')                     #setting y label
    axes[1].tick_params(rotation=90)                   #rotating the x-axis labels at 90*
    
    plt.suptitle(title, fontsize=20)                  #Setting the main title
    
DTITLE = "24 hours top losers and winners"
file =pd.read_csv('coinmarketcap_06122017.csv')

sub_plots(file,DTITLE)
```

# Task No 8:

## Checking the weekly Series too.¶

```python
def sub_plots_7d(volatility,title):
    figure, axes =plt.subplots(nrows=1,ncols=2,figsize=(8,6))
    

    volatility=volatility[['id','percent_change_7d']].dropna().sort_values(by='percent_change_7d')
    list_of_10=np.arange(10)
    
    #For 1st bar chart
    
    lossers=volatility.head(10)
    
    axes[0].bar(list_of_10,lossers.percent_change_7d) #plotting bar using id and % change
    axes[0].set_xticks(list_of_10)                     #setting xticks
    axes[0].set_xticklabels(lossers.id)                #now setting labels on x ticks
    axes[0].set_title("Losers",color='red')            #setting title
    axes[0].set_ylabel('% change')                     #setting y label
    axes[0].tick_params(rotation=90)                   #rotating the x-axis labels at 90*
    
    #For 2nd bar chart
    
    winners=volatility.tail(10)
    
    axes[1].bar(list_of_10,winners.percent_change_7d) #plotting bar using id and % change
    axes[1].set_xticks(list_of_10)                     #setting xticks
    axes[1].set_xticklabels(winners.id)                #now setting labels on x ticks
    axes[1].set_title("Winners",color='red')           #setting title
    axes[1].set_ylabel('% change')                     #setting y label
    axes[1].tick_params(rotation=90)                   #rotating the x-axis labels at 90*
    
    plt.suptitle(title, fontsize=20)                  #Setting the main title
    
    
DTITLE = "Weekly top lossers and winners"
volatility_7d =pd.read_csv('coinmarketcap_06122017.csv')

sub_plots_7d(volatility_7d,DTITLE)
```

# Task No 10:

## Most coins are tiny

```python
file=pd.read_csv('coinmarketcap_06122017.csv').dropna()
capital=file[['id','market_cap_usd']]
capital=capital.dropna()

def captal_volume(small_volume,large_volume):
    if large_volume>small_volume:
        return len(capital[(capital['market_cap_usd'] >= small_volume) & (capital['market_cap_usd'] < large_volume)])
    
biggest = captal_volume(300000000, 500000000)
micro = captal_volume(50000000, 300000000)
nano = captal_volume(0, 50000000)
print('biggest',biggest)
print('micro',micro)
print('nano',nano)

labels=[ "micro", "nano","biggest"]
plt.bar(np.arange(3),[biggest,micro,nano])
plt.xticks(np.arange(3),labels)
```
